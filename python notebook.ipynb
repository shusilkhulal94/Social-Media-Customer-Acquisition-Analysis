```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('/Users/shusilkhulal/Downloads/social_media_cleandataset.csv')
```

```python
corr = df[[
    'views',
    'clicks',
    'shares',
    'total_conversion',
    'revenue',
    'ad_spend',
    'roas',
    'ctr'
]].corr()
```

```python
plt.figure(figsize=(10, 8))

sns.heatmap(
    corr,
    annot=True,
    cmap='Blues',
    fmt='.2f'
)

plt.title("Social Media Metrics Correlation Matrix")
plt.tight_layout()
plt.show()
```
